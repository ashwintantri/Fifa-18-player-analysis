### TRYING SIMPLE LINEAR REGRESSION

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
reg = linear_model.LinearRegression()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
## plot learning curve









## Now interpret, overfitting, underfitting etc. from learning curve

## to be deleted, only for reference
i = 100
train_err = []
test_err = []
i_list = []
while(i <= len(X_train)):
    X_train_lc = X_train.iloc[0:i]
    y_train_lc = y_train.iloc[0:i]
    reg.fit(X_train_lc,y_train_lc)
    y_pred_test = reg.predict(X_test)
    y_pred_train = reg.predict(X_train_lc)
    train_err.append(mean_squared_error(y_train_lc,y_pred_train))
    test_err.append(mean_squared_error(y_pred_test,y_test))
    i_list.append(i)
    i = i + 500
plt.plot(i_list,train_err)    
plt.plot(i_list,test_err)


## The model is underfitting as more no. of samples is not improving the accuracy.
```
